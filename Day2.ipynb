```python
from random import randint

def roll_dice(n = 2):
    """
    摇骰子
    :param n: 骰子个数
    :return: n颗骰子数字之和
    """
    
    total = 0
    for _ in range(n):
        total += randint(1, 6)
    
    return total

def add(a = 0, b = 0, c = 0):
    return a + b + c

#在python程序中，很重要的一点是程序可以设置默认值
print(roll_dice())
print(roll_dice(3))
print(roll_dice(7))

print(add())
print(add(2, 3))
print(add(1))
print(add(1, 2, 3))

#竟然还可以这样不区分顺序！！这一点很有意思
print(add(c=50, a=100, b=200))
```

```python
#练习1:判断一个数是不是素数
import math
def isPrime(num):
    cell = int(math.sqrt(num)) + 1
    for i in range(2, cell):
        if num % i == 0:
            return False
    return True

isPrime(29)
```

```python
#练习2：判断一个数是不是回文数

#思路：把这个数字倒着写就好了
def is_palindrome(num):
    temp = num
    total = 0
    while temp > 0:
        total = total * 10 + temp % 10
        temp //= 10
    return temp == num
    
```

```python
#计算最小公倍数和最大公约数

def gcd(x, y):
    if y > x:
        x, y = y, x
    for i in range(y, 0, -1):
        if x % i == 0 and y % i == 0:
            return i
        #return后程序直接结束

def lcm(x, y):
    return x * y / gcd(x, y)
    
```

```python
#斐波那契数列的生成器
def fib(n):
    a = 0
    b = 1
    for i in range(n):
        a, b = b, a + b
        yield a

type(fib(20))
#可以看出生成器明显占用更小的内存，而且内存一致
print(sys.getsizeof(fib(40)))
print(sys.getsizeof(fib(80)))
print(sys.getsizeof(fib(100000)))
for val in fib(20):
    print(val)
    
        
```

```python
#使用字典
#字典是另一种可变容器模型，类似于我们生活中使用的字典，它可以存储任意类型对象
#与列表、集合不同的是，字典的每个元素都是由一个键和一个值组成的“键值对”，
#键和值通过冒号分开。下面的代码演示了如何定义和使用字典。

def main():
    animals = {'Romeo': 'Pig', 'KK': 'Elephant', 'Robin': 'Chicken'}
    print(animals['Romeo'])
    print(animals['KK'])
    
    for _ in animals:
        print("%s\tis%s" % (_, animals[_]))
        
    animals['Romeo'] = 'Dog'
    print(animals)
    animals['Sky'] = 'Cat'
    print(animals)
    #Error: print(animals[1])
    
    #删除字典的最后一个项目
    animals.popitem()
    #删除字典的指定项目
    animals.pop('KK')
    print(animals)
    
    #清空字典
    animals.clear()
    print(animals)
if __name__ == "__main__":
    main()
```

```python
#计算指定年月日是这一年的第几天

def isLeap(year):
    if year % 400 == 0:
        return True
    elif year % 4 == 0 and year % 100 != 0:
        return True
    else: return False

def which_day(year, month, date):
    days_of_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    if isLeap(year):
        days_of_month[1] = 29
    return sum(days_of_month[0:month - 1]) + date

def main():
    print(which_day(1980, 11, 28))
    print(which_day(1981, 12, 31))
    print(which_day(2018, 1, 1))
    print(which_day(2016, 3, 1))


if __name__ == '__main__':
    main()
```

```python
def main():
    num = int(input('Number of rows: '))
    yh = [[]] * num
    for row in range(len(yh)):
        yh[row] = [None] * (row + 1)
        for col in range(len(yh[row])):
            if col == 0 or col == row:
                yh[row][col] = 1
            else:
                yh[row][col] = yh[row - 1][col] + yh[row - 1][col - 1]
            print(yh[row][col], end='\t')
        print()


if __name__ == '__main__':
    main()
```

```python
import os


def print_board(board):
    print(board['TL'] + '|' + board['TM'] + '|' + board['TR'])
    print('-+-+-')
    print(board['ML'] + '|' + board['MM'] + '|' + board['MR'])
    print('-+-+-')
    print(board['BL'] + '|' + board['BM'] + '|' + board['BR'])


def main():
    init_board = {
        'TL': ' ', 'TM': ' ', 'TR': ' ',
        'ML': ' ', 'MM': ' ', 'MR': ' ',
        'BL': ' ', 'BM': ' ', 'BR': ' '
    }
    begin = True
    while begin:
        curr_board = init_board.copy()
        begin = False
        turn = 'x'
        counter = 0
        os.system('clear')
        print_board(curr_board)
        while counter < 9:
            move = input('轮到%s走棋, 请输入位置: ' % turn)
            if curr_board[move] == ' ':
                counter += 1
                curr_board[move] = turn
                if turn == 'x':
                    turn = 'o'
                else:
                    turn = 'x'
            os.system('clear')
            print_board(curr_board)
        choice = input('再玩一局?(yes|no)')
        begin = choice == 'yes'


if __name__ == '__main__':
    main()
```

```python
#综合案例：井字棋

#打印棋局
def print_board(board):
    print(board['TL'] + '|' + board['TM'] + '|' + board['TR'])
    print('-+-+-')
    print(board['ML'] + '|' + board['MM'] + '|' + board['MR'])
    print('-+-+-')
    print(board['BL'] + '|' + board['BM'] + '|' + board['BR'])
    
#判断棋局是否结束
def isOver(board, count):
    if count == 9:
        return True
    elif (board['TL'] == board['TM'] and board['TR'] == board['TM']) and (board['TL'] == 'o' or board['TL'] == 'x'):
        return True
    elif (board['ML'] == board['MM'] and board['MR'] == board['MM']) and (board['ML'] == 'o' or board['ML'] == 'x'):
        return True
    elif (board['BL'] == board['BM'] and board['BR'] == board['BM']) and (board['BL'] == 'o' or board['BL'] == 'x'):
        return True
    elif (board['TL'] == board['MM'] and board['MM'] == board['BR']) and (board['TL'] == 'o' or board['TL'] == 'x'):
        return True
    elif (board['TR'] == board['MM'] and board['BL'] == board['MM']) and (board['TR'] == 'o' or board['TR'] == 'x'):
        return True
    else: return False


def main():
    init_board = {'TL': ' ', 'TM': ' ', 'TR': ' ',
                  'ML': ' ', 'MM': ' ', 'MR': ' ',
                  'BL': ' ', 'BM': ' ', 'BR': ' '}
    count = 0
    turn = 'o'
    curr_board = init_board.copy()
    print_board(curr_board)
    while not(isOver(curr_board, count)):
        move = input("Now it is turn for " + turn + ':')
        curr_board[move] = turn
        print_board(curr_board)
        count += 1
        if isOver(curr_board, count):
            print(turn + " win the game!")
        elif turn == 'o':
            turn = 'x'
        else: turn = 'o'
    
if __name__ == '__main__':
    main()
```

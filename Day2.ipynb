```python
#ÁªÉ‰π†1:Âà§Êñ≠‰∏Ä‰∏™Êï∞ÊòØ‰∏çÊòØÁ¥†Êï∞
import math
def isPrime(num):
    cell = int(math.sqrt(num)) + 1
    for i in range(2, cell):
        if num % i == 0:
            return False
    return True

isPrime(29)
```

```python
#ÁªÉ‰π†2ÔºöÂà§Êñ≠‰∏Ä‰∏™Êï∞ÊòØ‰∏çÊòØÂõûÊñáÊï∞

#ÊÄùË∑ØÔºöÊääËøô‰∏™Êï∞Â≠óÂÄíÁùÄÂÜôÂ∞±Â•Ω‰∫Ü
def is_palindrome(num):
    temp = num
    total = 0
    while temp > 0:
        total = total * 10 + temp % 10
        temp //= 10
    return temp == num
    
```

```python
#ËÆ°ÁÆóÊúÄÂ∞èÂÖ¨ÂÄçÊï∞ÂíåÊúÄÂ§ßÂÖ¨Á∫¶Êï∞

def gcd(x, y):
    if y > x:
        x, y = y, x
    for i in range(y, 0, -1):
        if x % i == 0 and y % i == 0:
            return i
        #returnÂêéÁ®ãÂ∫èÁõ¥Êé•ÁªìÊùü

def lcm(x, y):
    return x * y / gcd(x, y)
    
```

```python
#ÊñêÊ≥¢ÈÇ£Â•ëÊï∞ÂàóÁöÑÁîüÊàêÂô®
def fib(n):
    a = 0
    b = 1
    for i in range(n):
        a, b = b, a + b
        yield a

type(fib(20))
#ÂèØ‰ª•ÁúãÂá∫ÁîüÊàêÂô®ÊòéÊòæÂç†Áî®Êõ¥Â∞èÁöÑÂÜÖÂ≠òÔºåËÄå‰∏îÂÜÖÂ≠ò‰∏ÄËá¥
print(sys.getsizeof(fib(40)))
print(sys.getsizeof(fib(80)))
print(sys.getsizeof(fib(100000)))
for val in fib(20):
    print(val)
    
        
```

```python
#ËÆ°ÁÆóÊåáÂÆöÂπ¥ÊúàÊó•ÊòØËøô‰∏ÄÂπ¥ÁöÑÁ¨¨Âá†Â§©

def isLeap(year):
    if year % 400 == 0:
        return True
    elif year % 4 == 0 and year % 100 != 0:
        return True
    else: return False

def which_day(year, month, date):
    days_of_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    if isLeap(year):
        days_of_month[1] = 29
    return sum(days_of_month[0:month - 1]) + date

def main():
    print(which_day(1980, 11, 28))
    print(which_day(1981, 12, 31))
    print(which_day(2018, 1, 1))
    print(which_day(2016, 3, 1))


if __name__ == '__main__':
    main()
```

```python
def main():
    num = int(input('Number of rows: '))
    yh = [[]] * num
    for row in range(len(yh)):
        yh[row] = [None] * (row + 1)
        for col in range(len(yh[row])):
            if col == 0 or col == row:
                yh[row][col] = 1
            else:
                yh[row][col] = yh[row - 1][col] + yh[row - 1][col - 1]
            print(yh[row][col], end='\t')
        print()


if __name__ == '__main__':
    main()
```

```python
import os


def print_board(board):
    print(board['TL'] + '|' + board['TM'] + '|' + board['TR'])
    print('-+-+-')
    print(board['ML'] + '|' + board['MM'] + '|' + board['MR'])
    print('-+-+-')
    print(board['BL'] + '|' + board['BM'] + '|' + board['BR'])


def main():
    init_board = {
        'TL': ' ', 'TM': ' ', 'TR': ' ',
        'ML': ' ', 'MM': ' ', 'MR': ' ',
        'BL': ' ', 'BM': ' ', 'BR': ' '
    }
    begin = True
    while begin:
        curr_board = init_board.copy()
        begin = False
        turn = 'x'
        counter = 0
        os.system('clear')
        print_board(curr_board)
        while counter < 9:
            move = input('ËΩÆÂà∞%sËµ∞Ê£ã, ËØ∑ËæìÂÖ•‰ΩçÁΩÆ: ' % turn)
            if curr_board[move] == ' ':
                counter += 1
                curr_board[move] = turn
                if turn == 'x':
                    turn = 'o'
                else:
                    turn = 'x'
            os.system('clear')
            print_board(curr_board)
        choice = input('ÂÜçÁé©‰∏ÄÂ±Ä?(yes|no)')
        begin = choice == 'yes'


if __name__ == '__main__':
    main()
```

```python
#ÁªºÂêàÊ°à‰æãÔºö‰∫ïÂ≠óÊ£ã

#ÊâìÂç∞Ê£ãÂ±Ä
def print_board(board):
    print(board['TL'] + '|' + board['TM'] + '|' + board['TR'])
    print('-+-+-')
    print(board['ML'] + '|' + board['MM'] + '|' + board['MR'])
    print('-+-+-')
    print(board['BL'] + '|' + board['BM'] + '|' + board['BR'])
    
#Âà§Êñ≠Ê£ãÂ±ÄÊòØÂê¶ÁªìÊùü
def isOver(board, count):
    if count == 9:
        return True
    elif (board['TL'] == board['TM'] and board['TR'] == board['TM']) and (board['TL'] == 'o' or board['TL'] == 'x'):
        return True
    elif (board['ML'] == board['MM'] and board['MR'] == board['MM']) and (board['ML'] == 'o' or board['ML'] == 'x'):
        return True
    elif (board['BL'] == board['BM'] and board['BR'] == board['BM']) and (board['BL'] == 'o' or board['BL'] == 'x'):
        return True
    elif (board['TL'] == board['MM'] and board['MM'] == board['BR']) and (board['TL'] == 'o' or board['TL'] == 'x'):
        return True
    elif (board['TR'] == board['MM'] and board['BL'] == board['MM']) and (board['TR'] == 'o' or board['TR'] == 'x'):
        return True
    else: return False


def main():
    init_board = {'TL': ' ', 'TM': ' ', 'TR': ' ',
                  'ML': ' ', 'MM': ' ', 'MR': ' ',
                  'BL': ' ', 'BM': ' ', 'BR': ' '}
    count = 0
    turn = 'o'
    curr_board = init_board.copy()
    print_board(curr_board)
    while not(isOver(curr_board, count)):
        move = input("Now it is turn for " + turn + ':')
        curr_board[move] = turn
        print_board(curr_board)
        count += 1
        if isOver(curr_board, count):
            print(turn + " win the game!")
        elif turn == 'o':
            turn = 'x'
        else: turn = 'o'
    
if __name__ == '__main__':
    main()
```
